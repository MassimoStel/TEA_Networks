```python
import teanets as tea

import time
import psutil
import os
import torch


text =     "Mathias is a clever student. He always delivers nice projects, rich and detailed. However, his friend Thomas is a terrible scholar. They are very different but also very much aligned in loving mathematics. Their favourite subjects are maths and physics. Mathias can do very well in his homeworks."

def get_memory_snapshot():
    process = psutil.Process(os.getpid())
    ram_mb = process.memory_info().rss / 1024 / 1024
    gpu_mb = None
    if torch.cuda.is_available():
        gpu_mb = torch.cuda.memory_allocated() / 1024 / 1024
    return ram_mb, gpu_mb


def run_timed(label, fn):
    ram_before, gpu_before = get_memory_snapshot()
    
    psutil.cpu_percent(interval=None)          # reset counter
    psutil.cpu_percent(percpu=True)            # reset per-core counter
    
    t_start = time.perf_counter()
    result = fn()
    elapsed = time.perf_counter() - t_start
    
    cpu_used = psutil.cpu_percent(interval=None)           # total since reset
    cpu_per_core = psutil.cpu_percent(percpu=True)         # per core since reset
    
    ram_after, gpu_after = get_memory_snapshot()

    print(f"\n[{label}]")
    print(f"  Time          : {elapsed:.3f}s")
    print(f"  CPU total     : {cpu_used:.1f}%")
    print(f"  CPU per core  : {[f'{c:.0f}%' for c in cpu_per_core]}")
    print(f"  RAM before    : {ram_before:.1f} MB")
    print(f"  RAM after     : {ram_after:.1f} MB")
    print(f"  RAM delta     : {ram_after - ram_before:+.1f} MB")
    if gpu_before is not None:
        print(f"  GPU before    : {gpu_before:.1f} MB")
        print(f"  GPU after     : {gpu_after:.1f} MB")
        print(f"  GPU delta     : {gpu_after - gpu_before:+.1f} MB")
    else:
        print(f"  GPU           : not available")

    return result, elapsed


# --- your loop ---
# for text in texts:
print("\n" + "="*60)
print(f"TEXT: {text[:60]}...")
print("="*60)

svo_with, t1 = run_timed(
    "fastcoref",
    lambda: tea.extract_svos_from_text(text, coref_solver="fastcoref", use_gpu=True)
)
svo_without, t2 = run_timed(
    "no coref",
    lambda: tea.extract_svos_from_text(text, coref_solver=None, use_gpu=True)
)
svo_stanza, t3 = run_timed(
    "stanza",
    lambda: tea.extract_svos_from_text(text, coref_solver="stanza", use_gpu=True)
)

print(f"\n--- Summary ---")
print(f"  fastcoref : {t1:.3f}s")
print(f"  no coref  : {t2:.3f}s")
print(f"  stanza    : {t3:.3f}s")
print(f"  fastest   : {'fastcoref' if t1 < t3 else 'stanza'}")

print("\nWITH coref (fastcoref):")
print(svo_with[svo_with["TEA"] == "Agent"]["Node 1"].unique())
print("\nWITHOUT coref:")
print(svo_without[svo_without["TEA"] == "Agent"]["Node 1"].unique())
print("\nWITH STANZA coref:")
print(svo_stanza[svo_stanza["TEA"] == "Agent"]["Node 1"].unique())
print("\n" + "-"*50)
```
