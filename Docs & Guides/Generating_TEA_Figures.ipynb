# Generating TEA Figures

This notebook generates the TEA network figures used in the README.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import teanets as tea
```

## 1. Framework overview (`tea_overview.jpg`)

Three sentences covering passive voice, active voice, and negative valence:

```python
text = (
    "Mark and Rose go to the park. "
    "The hacker destroyed the sensitive database."
    "The physician and the doctor were praised by the hospital."
)

svo = tea.extract_svos_from_text(text)
display(svo)
```

```python
tea.plot_svo_graph(svo, filename='../tea_overview.jpg')
```

## 2. Quickstart (`tea_quickstart.jpg`)

A single sentence with one agent performing two actions:

```python
text = """The physician and the doctor were praised by the hospital."""

svo = tea.extract_svos_from_text(text)
display(svo)
```

```python
tea.plot_svo_graph(svo, filename='../tea_quickstart_passive.jpg')
```

## 3. Multi-sentence example (`tea_multisentence.jpg`)

Two sentences with mixed valence — negative exposure and positive praise:

```python
text = (
    "The system detected an error. Finally, the loaf was sold."
    "The mayor praised the volunteers and thanked the community."
)

svo = tea.extract_svos_from_text(text)
display(svo)
```
